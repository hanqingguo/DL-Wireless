```python
import numpy as np
import matplotlib.pyplot as plt
```

In this note, trying to plot 16QAM analog wave.
![img/16-qam.png](img/16-qam.png)

As shown above, Suppose $f_{c}$ = 100

When trying to transmit 0011 with 16-QAM, then 

X(t) = $cos(2\pi f_{c}t) - 3sin(2\pi f_{c}t)$

```python
pi = np.pi
fc = np.power(10,2)
```

```python
t = np.linspace(0,10/fc,400)   # 10/fc 表示 10个周期, 在10个周期内取400个点
```

```python
phi = 2*pi*fc*t
```

```python
xt = np.cos(phi) - 3*np.sin(phi)
plt.plot(t, xt)
```

When trying to transmit 0111,

I = 1, Q = -1 

$a(t) = \sqrt{2}$, $\theta(t) = -\frac{\pi}{2}rad$

$X(t) = $cos(2\pi f_{c}t) - sin(2\pi f_{c}t)$$

```python
t2 = np.linspace(10/fc,20/fc,400)
phi2 = 2*pi*fc*t2
x3 = np.cos(phi2) - np.sin(phi2)
x_sum = np.concatenate((xt, x3), axis=0)
```

把发送0011 和 0111 的analog wave 连在一起，就是如下图的样子

```python
t_sum = np.linspace(0, 20/fc, 800)
plt.plot(t_sum, x_sum)
```
